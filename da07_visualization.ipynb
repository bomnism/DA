*  2개의 연속형 변수 간의 상관 관계: scatter plot

        * mpg ~ cylinder scatter
        * mpg ~ horsepower scatter
        * mpg ~ weight scatter
        * mpg ~ acceleration scatter 


```python
plt.scatter(x='displacement', y='mpg', data=mpg)
plt.xlabel('displacement')
plt.ylabel('mpg')

plt.show
```

```python
sns.scatterplot(x='displacement', y='mpg', data=mpg,  s=40)
plt.show()  
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 6))            # 2x2 생성 
# print(ax)
# plt.show()

ax[0, 0].scatter(y='mpg', x='cylinders', data = mpg)
ax[0, 0].set_title('cylinders')
ax[0, 1].scatter(y='mpg', x='horsepower', data = mpg)
ax[0, 1].set_title('horsepower')
ax[1, 0].scatter(y='mpg', x='weight', data = mpg)
ax[1, 0].set_title('weight')
ax[1, 1].scatter(y='mpg', x='acceleration', data = mpg)
ax[1, 1].set_title('acceleration')
```

* cylinder별 mpg 평균 막대 그래프

```python
sns.barplot(x = 'cylinders', y='mpg', data = mpg)
plt.show()
```

```python
 cylinders = mpg['cylinders'].unique()
 cylinders.sort()       # 배열을 정렬된 상태로 변환
 print(cylinders)
```

```python
mean_mpg = []       # cylinder 개수별 mpg의 평균을 저장할 리스트 
for c in cylinders:
    mu = mpg[mpg['cylinders'] == c]['mpg'].mean()
    # print(c, mu)
    mean_mpg.append(mu)

s = pd.Series(data=mean_mpg, index = cylinders, name='mean_mpg')

s
```

```python
plt.bar(x=s.index, height=s.values)
plt.xlabel('cylinders')
plt.ylabel('mpg')
plt.show()
```


#iris 데이터 셋
* 데이터 셋 탐색
* 품종을 제외한 4개 변수의 boxplot
* 품종별로 Subplot의 컬럼을 구분하고, 숫자 타입 변수들의 boxplot을 각 품종별로 시각화
* petal_width ~ petal_length scatter plot. 품종별로 점의 색깔을 다르게.
* 모든 연속형 변수들 간의 상관 관계 scatter plot - Pair Plot

```python
iris = sns.load_dataset('iris')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

sns.barplot(x='origin', y = 'mpg', data = mpg, ax=ax[0])
sns.barplot(x = 'cylinders', y = 'mpg', data=mpg, ax=ax[1])

plt.show()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris['species'].value_counts()
```

```python
# 품종을 제외한 4개 변수의 boxplot
sns.boxplot(data=iris)    # 데이터프레임의 모든 숫자 타입 변수들의 boxplot을 그려줌. 
plt.show()                          # 모든 변수들에 대한 boxplot
```

```python
vars = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
data = [iris[x] for x in vars]
plt.boxplot(x =data, labels = vars)
plt.show()
```

```python
# 품종별로 Subplot의 컬럼을 구분하고, 숫자 타입 변수들의 boxplot을 각 품종별로 시각화
sns.catplot(data=iris, col='species', kind='box')
plt.show()
```

```python
# petal_width ~ petal_length scatter plot. 품종별로 점의 색깔을 다르게.
sns.scatterplot(data = iris, x = 'petal_length', y='petal_width', hue = 'species')
```

```python
# 모든 연속형 변수들 간의 상관 관계 scatter plot - Pair Plot  /  densityplot(막대그래프 연결한 모양)
sns.pairplot(data=iris, hue='species')
plt.show()
```
