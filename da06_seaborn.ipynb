```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

tips 데이터 프레임 로드 

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

# Box Plot

```python
# total_bill box plot
sns.boxplot(y='total_bill', data=tips)
# sns.boxplot(y=tips['total_bill'])과 동일 
plt.show()
```

```python
sns.boxplot(x='total_bill', data=tips)
plt.show()
```

성별 tip의 boxplot

```python
sns.boxplot(x='sex', y='tip', data=tips)
plt.show()
```

요일별 tip의 boxplot

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.show()
```

* subplot 2개(1x2) 생성
* 성별 tip boxplot, 흡연여부별 tip boxplot

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
# ax: Subplot 객체들을 가지고 있는 1차원 배열(ndarray)

sns.boxplot(x='sex', y='tip', data=tips, ax=ax[0])
sns.boxplot(x='smoker', y='tip', data=tips, ax=ax[1])

plt.show()
```

## Histogram

```python
#  total_bill histogram
sns.histplot(x='total_bill', data=tips, bins=10)
plt.show()
```

```python
sns.histplot(y='total_bill', data=tips, bins=10)
plt.show()
```

## Bar plot

## `countplot()`
카테고리 타입 변수 빈도수(count)를 막대로 표현한 그래프 .  변수 하나로  

```python
#  남성/여성 수 - 세로 막대 그래프
sns.countplot(x='sex', data=tips)
plt.show()
```

```python
#  흡연자/비흡연자 수 - 가로 막대 그래프 
sns.countplot(y = 'smoker', data = tips)
plt.show()
```

```python
# 요일별 성별 빈도수 
sns.countplot(x = 'day', data = tips, hue='sex')
plt.show()
```

```python
#  성별, 흡연자별 빈도수 
# sns.countplot(x='smoker', data = tips, hue='sex)
sns.countplot(x='sex', data = tips, hue='smoker', palette=['dodgerblue', 'indianred'])
plt.show()
```

## `barplot()`

그룹별 변수의 요약 통계량(예: 평균) / 카테고리별, 연속형 변수의 요약 통계량을 함께 보는 것 

```python
# 성별 tip의 평균
print(tips[tips['sex'] == 'Male'] ['tip'].mean())
print(tips[tips['sex'] == 'Female'] ['tip'].mean())
```

```python
sns.barplot(x='sex', y ='tip', data=tips, estimator=np.max)
plt.show()
```

```python
#  요일별 tip의 평균 가로 막대 그래프 
sns.barplot(x='tip', y='day', data=tips)
plt.show()
```

```python
# 요일별 성별 tip의 평균 세로 막대 그래프 
sns.barplot(x='day', y ='tip', hue='sex', data=tips)
plt.show()
```

```python
sns.barplot(x='sex', y = 'tip', hue='day', data=tips)
plt.show()
```

# Scatter Plot

```python
# tip ~ total_bill
sns.scatterplot(x='total_bill', y='tip', data=tips, s=70)
plt.show()      # 양의 상관관계 
```

```python
# regplot: regression plot. 산점도 그래프 + 선형 회귀 직선 
sns.regplot(x='total_bill' ,  y='tip', data=tips, scatter_kws={'s' : 10})
plt.show()
```

```python
#  jointplot: 산점도 그래프 + 히스토그램 
sns.jointplot(x='total_bill', y = 'tip', data = tips)
plt.show()
```

```python
# tip ~ total_bill 산점도 + 점의 색깔을 성별에 따라 다르게 
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips, size='size')
plt.show()  
```

```python
# tip ~ total_bill 산점도 + 점의 색깔을 성별에 따라 다르게 + 점의 크기를 인원수에 따라 다르게 
sns.scatterplot(x='total_bill', y='tip', hue='sex', size = 'size', data=tips, s=70)
plt.show()  
```

```python
# tip ~ total_bill 산점도 + 점의 색깔을 성별에 따라 다르게 + 점의 모양을 시간에 따라 다르게 
sns.scatterplot(x='total_bill', y='tip', hue='sex', style = 'time', data=tips,  s=70)
plt.show()  
```

# Category Plot
카테고리별로 Subplot을 생성하고 통계를 계산해서, 모든 Subplot에 같은 종류의 그래프들로 시각화. 

```python
# 요일별(x축) tip의 평균(y축) 막대 그래프  
# 흡연여부별(row) subplot 생성, 시간별(column) subplot을 생성 
sns.catplot(x='day', y='tip', data=tips, 
                    kind = 'bar',        # kind: subplot에 그릴 그래프 종류 
                    row='smoker',   # row: row subplot을 나누는 기준
                    col='time')          # col: column subplot을 나누는 기준 
plt.show() 
```

```python
tips[(tips.smoker == 'No') & (tips.time == 'Lunch') & (tips.day == 'Fri')]      # 한명만 있기 때문에 표준편차가 없음(잣대기가 없는 이유)
```

```python
tips[(tips.smoker == 'No') & (tips.time == 'Dinner') & (tips.day == 'Fri')]      # 세명이 존재, 다른거에 비해 커서 잣대기가 길다 
```
